```python
import re
text = " ".join(df['PlayerLine'])
text = text.lower()
text = re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
print(text[:500])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
vocab_size = len(tokenizer.word_index) + 1
print(vocab_size)
```

```python
sequence = tokenizer.texts_to_sequences([text])[0]
print(sequence[:20])
```

```python
import numpy as np

sequences = []

seq_length = 5

# Ensure sequence is defined and long enough before trying to create sub-sequences
if 'sequence' in locals() and len(sequence) >= seq_length:
    for i in range(seq_length, len(sequence)):
        seq = sequence[i-seq_length:i+1]
        sequences.append(seq)

sequences = np.array(sequences)
```

```python
sequences[:5]
```

```python
import numpy as np

# Check if sequences is not empty before slicing
if sequences.size > 0:
    X = sequences[:, :-1]
    y = sequences[:, -1]
else:
    X = np.array([]) # Initialize X and y as empty numpy arrays
    y = np.array([])
    print("Warning: 'sequences' array is empty, X and y are initialized as empty arrays.")
```

```python
print(X.shape)
print(y.shape)
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, num_classes=vocab_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

model = Sequential()

model.add(Embedding(vocab_size, 100, input_length=seq_length))
model.add(SimpleRNN(150))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
model.fit(
    X,
    y,
    epochs=5,
    batch_size=64
)
```
